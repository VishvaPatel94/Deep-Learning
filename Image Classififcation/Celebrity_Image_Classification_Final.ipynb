# Data Augumentation

```python
datagen = ImageDataGenerator(
        rotation_range=45,
        width_shift_range=0.2,
        height_shift_range=0.2,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True,
        fill_mode='nearest')
```

```python
import os 
import math
import glob 
img_dir = "Faces/Data/"  
data_path = os.path.join(img_dir,'*/') 
folders = glob.glob(data_path) 
labels=[]
training_files=[]
validation_files=[]
test_files=[]
images={}
for folder in folders: 
    img_dir = folder  
    data_path = os.path.join(img_dir,'*g') 
    files = glob.glob(data_path) 
    label=path.normpath(img_dir).split('\\')[-1]
    labels.append(label)
    num_files=len(files)
    print(img_dir)
    default_path=os.path.join(img_dir,'default.jpg')
    images[label]=default_path
    training_files.extend(files[:math.floor(0.7*num_files)])
    validation_files.extend(files[math.floor(0.7*num_files):math.floor(0.85*num_files)])
    test_files.extend(files[math.floor(0.85*num_files):])    
    for f1 in files[:math.floor(0.7*num_files)]: 
        img = load_img(f1)  # this is a PIL image
        img_name=path.normpath(f1).split('.')[1]
        print(img_name)
        x = img_to_array(img)  
        x = x.reshape((1,) + x.shape)  

        i = 0
        print(img_dir)
        for batch in datagen.flow(x, batch_size=1, save_to_dir=img_dir, save_prefix='prefix'+img_name, save_format='jpg'):
            i += 1
            if i > 20:
                break  
    data_path = os.path.join(img_dir,'*g')
    files = glob.glob(data_path) 
    da_files=[s for s in files if "prefix" in s]
    training_files.extend(da_files)
```

# Data Preperation

```python
IMG_SIZE = 100
```

```python
import cv2
import os
import numpy as np
from random import shuffle
from tqdm import tqdm
import math
```

```python
def label_img(img):
    word_label = path.normpath(img).split('.')[0].split('\\')[-2]
    index=labels.index(word_label)
    output=np.zeros(len(labels), dtype = int)
    output[index]=1
    return output
```

```python
def create_train_data():
    # Creating an empty list where we should store the training data
    # after a little preprocessing of the data
    training_data = []
  
    for img in training_files:
        # labeling the images
        print(img)
        label = label_img(img)
  
        # loading the image from the path and then converting them into
        # grayscale for easier covnet prob
        img = cv2.imread(img, cv2.IMREAD_GRAYSCALE)
  
        # resizing the image for processing them in the covnet
        img = cv2.resize(img, (IMG_SIZE, IMG_SIZE))
  
        # final step-forming the training data list with numpy array of the images
        training_data.append([np.array(img), np.array(label)])
  
    # shuffling of the training data to preserve the random state of our data
    shuffle(training_data)
    return training_data
```

```python
def process_validation_data(files):
    val_data = []
    for img in files:
        label = label_img(img)
        img = cv2.imread(img, cv2.IMREAD_GRAYSCALE)
        img = cv2.resize(img, (IMG_SIZE, IMG_SIZE))
        val_data.append([np.array(img), label])
          
    shuffle(val_data)
    np.save('val_data.npy', val_data)
    return val_data
```

```python
train_data = create_train_data()
```

```python
val_data = process_validation_data(validation_files)
```

```python
test_data=process_validation_data(test_files)
```

# Model Building

```python
#!pip install tflearn
```

```python
# Importing the required libraries
import tflearn
from tflearn.layers.conv import conv_2d, max_pool_2d
from tflearn.layers.core import input_data, dropout, fully_connected
from tflearn.layers.estimator import regression
  
import tensorflow as tf
```

```python
from tensorflow.keras import datasets, layers, models

tf.compat.v1.reset_default_graph()

model = models.Sequential()
model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(IMG_SIZE, IMG_SIZE, 1)))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(128, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(32, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Flatten())
model.add(layers.Dense(1024, activation='relu'))
model.add(layers.Dense(10, activation = 'sigmoid'))
```

```python
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3),
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.summary()
```

```python
'''Setting up the features and labels'''
# X-Features & Y-Labels
  
X = np.array([i[0] for i in train_data]).reshape(-1, IMG_SIZE, IMG_SIZE, 1)/255
Y = np.array([i[1] for i in train_data])
val_x = np.array([i[0] for i in val_data]).reshape(-1, IMG_SIZE, IMG_SIZE, 1)/255
val_y = np.array([i[1] for i in val_data])
```

```python
history = model.fit(X,Y,epochs=30, 
                    validation_data=(val_x,val_y))
```

# Testing

```python
test_x = np.array([i[0] for i in test_data]).reshape(-1, IMG_SIZE, IMG_SIZE, 1)/255
test_y = np.array([i[1] for i in test_data])
```

```python
test_acc = model.evaluate(test_x,test_y)
```

```python
def predict_face(n):
    filenames=random.sample(test_files, n)
    for img in filenames:
        im = Image.open(img)
        plt.rcParams["figure.figsize"] = [7.50, 3.50]
        plt.rcParams["figure.autolayout"] = True
        
            
        img = cv2.imread(img, cv2.IMREAD_GRAYSCALE)
        img = cv2.resize(img, (IMG_SIZE, IMG_SIZE))
        data = img.reshape(IMG_SIZE, IMG_SIZE, 1)/255
        model_out = model.predict(tf.expand_dims(data, 0),steps=1)[0]
        str_label =labels[np.argmax(model_out)]
        plt.subplot(1, 2, 1)
        plt.imshow(np.asarray(im))
        plt.axis('off')
        plt.subplot(1, 2, 2)
        plt.imshow(np.asarray(Image.open(images[str_label])))
        plt.axis('off')
        plt.show()
        print(model_out)
        print("This image most likely belongs to {} with a {:.2f} percent confidence."
              .format(str_label, 100 * np.max(model_out)))
```

```python
def sort_index(lst, rev=True):
    index = range(len(lst))
    s = sorted(index, reverse=rev, key=lambda i: lst[i])
    return s
```
